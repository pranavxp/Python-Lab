```python
import numpy as np
import matplotlib.pyplot as plt
x=np.linspace(0,7)
y=x**4+5
plt.plot(x,y)
plt.xlabel('x')
plt.ylabel('y')
plt.title('y=x**4+5')
plt.show
```

```python
#normal graph
import numpy as np
import matplotlib.pyplot as plt
x=np.linspace(-7,7)
y=np.exp(-x**2/2)/np.sqrt(2*np.pi)
plt.plot(x,y)
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

```python

#stem graph
import numpy as np
import matplotlib.pyplot as plt
x=['A','B','C','D','E','F','G']
y=['0','1','2','3','4','5','6']
plt.stem(x,y)
plt.xlabel('x')
plt.ylabel('y')
plt.title('stem graph')
plt.show()
```

```python
#bar graph
import numpy as np
import matplotlib.pyplot as plt
x=['A','B','C','D','E','F','G']
y=['0','1','2','3','4','5','6']
plt.bar(x,y)
plt.xlabel('x')
plt.ylabel('y')
plt.title('bar graph')
plt.show()
```

```python
import numpy as np
from matplotlib import pyplot as plt
t = np.linspace(0,4*np.pi,1000)
vin = np.sin(t)
vout = np.abs(vin)
plt.plot(t,vin,label="Input voltage")
plt.xlabel("Time")
plt.ylabel("Voltage")
plt.title("Full wave rectifier Input")
plt.show()

plt.plot(t,vout,label="Output voltage")
plt.xlabel("Time")
plt.ylabel("Voltage")
plt.title("Full wave rectifier Output")
plt.show()
```
