```python
from numpy import random
import matplotlib.pyplot as plt
import seaborn as sns
sns.distplot(random.binomial(n=1000,p=0.01,size=1000),hist = False)
plt.title('Binomial')
plt.show()
```

```python
from numpy import random
import matplotlib.pyplot as plt
import seaborn as sns
sns.distplot(random.poisson(lam=10,size=1000),hist = False)
plt.title('Poisson')
plt.show()
```
